```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Read the Parquet file
df = pd.read_parquet('/kaggle/input/child-mind-institute-problematic-internet-new/series_train.parquet/id=00115b9f')
print("Original DataFrame:")
print(df.head())


# Step 2: Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Step 3: Fit the scaler to the data and transform it
df_scaled = pd.DataFrame(scaler.fit_transform(df.select_dtypes(include=[np.number])), columns=df.select_dtypes(include=[np.number]).columns)
print("Scaled DataFrame:")
print(df_scaled.head())



# Step 3: Plot the Data in a Loop
columns = df_scaled.columns
for i in range(len(columns)):
    for j in range(i + 1, len(columns)):
        plt.figure(figsize=(10, 6))
        sns.scatterplot(data=df_scaled, x=columns[i], y=columns[j])
        plt.title(f"Scatter Plot of {columns[i]} vs {columns[j]}")
        plt.xlabel(columns[i])
        plt.ylabel(columns[j])
        plt.show()
```
